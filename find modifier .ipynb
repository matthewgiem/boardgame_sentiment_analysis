```python
df = pd.read_csv('data.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df['type'] = 0
df.loc[df.time>0, 'type'] = 1
df['normilization_divisor'] = 0
df['normilized_percent'] = 0
df['mean'] = 0
df['median'] = 0
df['mode'] = 0
```

```python
test = df[(df['word'].isin(['hate', 'fun']))]
```

```python
def write_statistics(series, df):
    numbers = []
    total = 0
    for x in series:
        if not math.isnan(x):
            numbers.append(x)
            total += x
    normilization_divisor = total
    df.loc[(df.word == series.name) & (df.type == 1),'normilization_divisor'] = normilization_divisor
```

```python
%%time
test.groupby(test.word).percent.apply(write_statistics, df=test)
```
