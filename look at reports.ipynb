```python
import pandas as pd
import glob
import math
from decimal import Decimal
```

```python
df = pd.DataFrame()
for report in glob.glob('new_report*'):
    if df.empty:
        df = pd.read_csv(report)
    else:
        df = pd.concat([pd.read_csv(report), df])
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'average/std': 'average_std'}, inplace=True)
```

```python
df.to_csv('data.csv')
```

```python
best_dict = dict(zip(df[['word', 'average_std']].dropna().loc[df.average_std.dropna()<1]['word'], df[['word', 'average_std']].dropna().loc[df.average_std.dropna()<1]['average_std']))
```

```python
sortedDict=sorted(best_dict.values())
sorted_dict = dict()
for sortedKey in sortedDict:
    for key, value in best_dict.items():
        if value==sortedKey:
            sorted_dict[key]=value
```

```python
def find_higest_power(series):
    higest = 100
    for x in series:
        if not math.isnan(x):
            if higest > int(str('%.2E' % Decimal(x)).split('E')[1][1:3]):
                if 0 != int(str('%.2E' % Decimal(x)).split('E')[1][1:3]):
                    higest = int(str('%.2E' % Decimal(x)).split('E')[1][1:3])
    how_many_are_zero = 0
    for x in series:
        if not math.isnan(x):
            if 0 == int(str('%.2E' % Decimal(x)).split('E')[1][1:3]):
                how_many_are_zero += 1
    if how_many_are_zero > 5:
        print(series.name)
#     return higest, how_many_are_zero
```

```python
test_df = df.copy()
```

```python
test_df.groupby('word').percent.apply(find_higest_power)
```
