```python
def length(var):
    return len(var.split(' '))
def find(var):
    return var.split(' ')
def word(var, good_word, bad_word):
    if good_word in [x.lower() for x in var.split(' ')]:
        return [8,9,10]
    if bad_word in [x.lower() for x in var.split(' ')]:
        return [1,2,3,4]
    else:
        return [5,6,7]
def check(a, b):
    if a in b:
        return True
    else:
        return False
    
df_small['simple_rating'] = df_small.apply(lambda row: word(row['comment'], 'great', 'terrible'), axis=1)

df_small['value'] = df_small.apply(lambda row: check(row['rating'], row['simple_rating']), axis=1)

df_small.groupby('value').count()

print('{}%'.format(round(df_small.loc[df_small.value == True].count()['rating']/df_small.count()['rating']*100,2)))
```

```python
bad = [1,2,3,4]
nuetral = [5,6,7]
good = [8,9,10]

def better(comment, rating, good_word, bad_word):
    prediction = None
    if good_word in [x.lower() for x in comment.split(' ')]:
        prediction = good
    elif bad_word in [x.lower() for x in comment.split(' ')]:
        prediction = bad
    else:
        prediction = nuetral
    if rating in prediction:
        return True
    else:
        return False
    
def prediction(df, good_word, bad_word):
    df['prediction'] = df.apply(lambda row: better(row['comment'], row['rating'], good_word, bad_word), axis=1)
    string = '{}'.format(round(df.loc[df.prediction == True].count()['rating']/df.count()['rating']*100,2))
    df.drop(columns=['prediction'], inplace=True)
    return float(string)

    
    
```

```python
prediction(df_small,'great', 'terrible')
```

```python
def loop(df, good_words, bad_words):
    program_starts = time.time()    
    best_combo = [0, '', '']
    for x in good_words:
        for y in bad_words:
            percentage = [prediction(df, x, y), x, y]
            if best_combo[0] < percentage[0]:
                best_combo = percentage
            now = time.time()
            print("It has been {0} seconds since the loop started".format(now - program_starts))
            print(best_combo)
    return best_combo
            
```

```python
good_words = ['love', 'fun', 'good', 'great']
bad_words = ['bad', 'board', 'boring', 'hate', 'never', 'broken', 'worst', 'long']
loop(df, good_words, bad_words)
```
