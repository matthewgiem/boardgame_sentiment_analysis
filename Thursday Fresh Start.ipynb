```python
%%time
df_large = pd.read_csv('../data/large_sample.csv')
df_large.drop(columns=['Unnamed: 0'], inplace=True)
df_large.rating = np.ceil(df_large.rating)
```

```python
def remove_punctuation(comment):
    table = str.maketrans(dict.fromkeys('@#$%^&*()_+=~`[{}]|\:;"<,>."/'))
    return comment.translate(table)
```

```python
def replace_exclimation(comment):
    return comment.replace('!', ' <!> ').replace('?', ' <?> ')
```

```python
%%time
df_large['remove_punctuation'] = df_large.comment.apply(lambda x: remove_punctuation(x))
df_large.drop(columns=['comment'], inplace=True)
df_large.rename(columns={'remove_punctuation':'comment'}, inplace=True)
df_large['replace_punctuation'] = df_large.comment.apply(lambda x: replace_exclimation(x))
df_large.drop(columns=['comment'], inplace=True)
df_large.rename(columns={'replace_punctuation':'comment'}, inplace=True)
```

```python
def make_it_lower(comment):
    new_comment = []
    for x in comment.split(' '):
        new_comment.append(x.lower())
    return ' '.join(new_comment)
```

```python
%%time
df_large['lower_comment'] = df_large.comment.apply(lambda x: make_it_lower(x))
df_large.drop(columns=['comment'], inplace=True)
df_large.rename(columns={'lower_comment':'comment'}, inplace=True)
```

```python
def word_count(comment, word, number):
    if word in comment.split(' '):
        if word in number:
            number[word] += comment.split(' ').count(word)
        else:
            number[word] = comment.split(' ').count(word)
```

```python
%%time
def make_report(w, z, dictionary, dataframe):
    dict_list = dictionary
    testing_df = dataframe.copy()
    labels = ['word', 'average/std', 'percent_average', 'percent_std', 'rating', 'word_count', 'divisor', 'percent', 'time']
    to_return = pd.DataFrame(columns=labels)
    for word in dict_list[int(w):int(z)]:
        start_time = tm.time()
        to_append = []
        std = 0
        average = 0
        for y in range(1,11):
            number = dict()
            testing_df.loc[testing_df.rating == y].comment.apply(word_count, word=word[0], number=number)
            divisor = testing_df.loc[testing_df.rating == y].comment.count()
            if word[0] in number:
                percent = number[word[0]]/divisor
            else:
                percent = 0
                number[word[0]] = 0
            to_append.append([word[0], None, None, None, y, number[word[0]], divisor, percent, None])
        for_array = []
        for value in to_append:
            for_array.append(value[7])
        x = np.array(for_array)
        array1 = [[word[0], (np.sum(x)/10)/(np.std(x)), np.sum(x)/10, np.std(x), None, None, None, None, tm.time()-start_time]]
        for x in to_append:
            array1.append(x)
        inbetween_df = pd.DataFrame.from_records(array1, columns=labels)
        to_return = pd.concat([to_return, inbetween_df])
    return to_return
```

```python
def write_csv(low, high, df):
    df.to_csv('new_report_for_{}_to_{}.csv'.format(low, high))
```

```python
def loop_write_report(dictionary, dataframe, number, count):
    num = int(number)
    size = int(count)
    first = 0
    while first < num - 1:
        if first + size < num - 2:
            report = make_report(first, first + size, dictionary, dataframe)
            write_csv(first, first + size, report)
            print('dataframe written from {} to {}'.format(first, first + size))
            first = first + size + 1
        else:
            report = make_report(first, num - 1, dictionary, dataframe)
            write_csv(first, num - 1, report)
            print('dataframe written from {} to {}'.format(first, num - 1))
            print('last dataframe')
            break
```
