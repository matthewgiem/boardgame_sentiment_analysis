```python
import pandas as pd
```

```python
df = pd.read_csv('normalized_data.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
test = df.loc[df.word={'hate', 'love'}]
```

```python
def write_value(series, df):
    value = 0
    counter = 0
    for x in series:
#         print(counter)
        if counter == 0:  
            value = x
#             print('the value for {} being wirtten is {}'.format(series.name, value))
        else:
            df.loc[series.index[counter], 'average_std'] = value
#             print(series.index[counter])
#             print(value)
        counter += 1
```

```python
%%time
df.groupby('word').average_std.apply(write_value, df=df)
```
